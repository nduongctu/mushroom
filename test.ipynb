```python
from sklearn.preprocessing import LabelEncoder

# Tạo một đối tượng LabelEncoder
le = LabelEncoder()

# Áp dụng mã hóa cho mỗi cột trong DataFrame và in chữ cái tương ứng với số đã chuyển đổi
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
for column in df.columns:
    df[column] = le.fit_transform(df[column])
    print(f"Column {column}:")
    print("Encoded Labels:", le.classes_)
    print("Encoded Values:", df[column].unique())
    print()

# Chia dữ liệu thành features (X) và target (y)
X = df.iloc[:, 1:]
y = df.iloc[:, 0]
print(df.head(1))
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, f1_score, recall_score, precision_score

# Số lần chạy
num_runs = 10

# List để lưu trữ độ chính xác, F1-score, recall và precision
accuracy_scores = []
f1_scores = []
recall_scores = []
precision_scores = []

# Vòng lặp để chạy mô hình và tính toán độ chính xác, F1-score, recall và precision
for i in range(num_runs):
    # Chia tập dữ liệu thành tập huấn luyện và tập kiểm tra
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

    # Khởi tạo mô hình KNN
    knn = KNeighborsClassifier(n_neighbors=10)

    # Huấn luyện mô hình
    knn.fit(X_train, y_train)

    # Dự đoán nhãn của dữ liệu kiểm tra
    y_pred = knn.predict(X_test)

    # Tính toán độ chính xác của mô hình
    accuracy_knn = accuracy_score(y_test, y_pred)
    accuracy_scores.append(accuracy_knn)
    print("Độ chính xác lần chạy thứ", i+1, ":", accuracy_knn)

    # Tính toán F1-score của mô hình
    f1_knn = f1_score(y_test, y_pred)
    f1_scores.append(f1_knn)
    print("F1-score lần chạy thứ", i+1, ":", f1_knn)

    # Tính toán độ phủ của mô hình
    recall_knn = recall_score(y_test, y_pred)
    recall_scores.append(recall_knn)
    print("Độ phủ lần chạy thứ", i+1, ":", recall_knn)

    # Tính toán độ chính xác ( precision) của mô hình
    precision_knn = precision_score(y_test, y_pred)
    precision_scores.append(precision_knn)
    print("Precision lần chạy thứ", i+1, ":", precision_knn)

# Độ chính xác trung bình
average_accuracy_knn = np.mean(accuracy_scores)
print("Độ chính xác trung bình sau", num_runs, "lần chạy:", average_accuracy_knn)

# F1-score trung bình
average_f1_knn = np.mean(f1_scores)
print("F1-score trung bình sau", num_runs, "lần chạy:", average_f1_knn)

# Độ phủ trung bình
average_recall_knn = np.mean(recall_scores)
print("Độ phủ trung bình sau", num_runs, "lần chạy:", average_recall_knn)

# Độ chính xác trung bình
average_precision_knn = np.mean(precision_scores)
print("Precision trung bình sau", num_runs, "lần chạy:", average_precision_knn)
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, f1_score, recall_score, precision_score

# Số lần chạy
num_runs = 10

# List để lưu trữ độ chính xác, F1-score, recall và precision
accuracy_scores = []
f1_scores = []
recall_scores = []
precision_scores = []

# Vòng lặp để chạy mô hình và tính toán độ chính xác, F1-score, recall và precision
for i in range(num_runs):
    # Chia tập dữ liệu thành tập huấn luyện và tập kiểm tra
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

    # Khởi tạo mô hình Random Forest
    rf = RandomForestClassifier(max_depth=10,n_estimators=50,criterion="gini")

    # Huấn luyện mô hình
    rf.fit(X_train, y_train)

    # Dự đoán nhãn của dữ liệu kiểm tra
    y_pred = rf.predict(X_test)

    # Tính toán độ chính xác của mô hình
    accuracy_rf = accuracy_score(y_test, y_pred)
    accuracy_scores.append(accuracy_rf)
    print("Độ chính xác lần chạy thứ", i+1, ":", accuracy_rf)

    # Tính toán F1-score của mô hình
    f1_rf = f1_score(y_test, y_pred)
    f1_scores.append(f1_rf)
    print("F1-score lần chạy thứ", i+1, ":", f1_rf)

    # Tính toán độ phủ của mô hình
    recall_rf = recall_score(y_test, y_pred)
    recall_scores.append(recall_rf)
    print("Độ phủ lần chạy thứ", i+1, ":", recall_rf)

    # Tính toán độ chính xác (precision) của mô hình
    precision_rf = precision_score(y_test, y_pred)
    precision_scores.append(precision_rf)
    print("Precision lần chạy thứ", i+1, ":", precision_rf)

# Độ chính xác trung bình
average_accuracy_rf = np.mean(accuracy_scores)
print("Độ chính xác trung bình sau", num_runs, "lần chạy:", average_accuracy_rf)

# F1-score trung bình
average_f1_rf = np.mean(f1_scores)
print("F1-score trung bình sau", num_runs, "lần chạy:", average_f1_rf)

# Độ phủ trung bình
average_recall_rf = np.mean(recall_scores)
print("Độ phủ trung bình sau", num_runs, "lần chạy:", average_recall_rf)

# Độ chính xác trung bình
average_precision_rf = np.mean(precision_scores)
print("Precision trung bình sau", num_runs, "lần chạy:", average_precision_rf)
```

```python

# Độ chính xác, recall, precision và F1-score của từng mô hình (trong phần trăm)
accuracy_scores = [average_accuracy_knn * 100, average_accuracy_nb * 100, average_accuracy_dt * 100, average_accuracy_rf * 100]
recall_scores = [average_recall_knn * 100, average_recall_nb * 100, average_recall_dt * 100, average_recall_rf * 100]
precision_scores = [average_precision_knn * 100, average_precision_nb * 100, average_precision_dt * 100, average_precision_rf * 100]
f1_scores = [average_f1_knn * 100, average_f1_nb * 100, average_f1_dt * 100, average_f1_rf * 100]

# Tên của từng mô hình
models = ['KNN', 'Naive Bayes', 'Decision Tree', 'Random Forest']

# Vẽ biểu đồ
plt.figure(figsize=(10, 10))

# Biểu đồ độ chính xác (trên)
plt.subplot(2, 2, 1)
plt.bar(models, accuracy_scores,width=0.5)
plt.title('Độ chính xác trung bình của các mô hình')
plt.ylabel('Accuracy (%)')
plt.ylim([0, 100])

# Ghi chú các giá trị phần trăm lên biểu đồ
for i, v in enumerate(accuracy_scores):
    plt.annotate(f'{v:.2f}%', xy=(i, v), ha='center', va='top')

# Biểu đồ độ phủ (trên)
plt.subplot(2, 2, 2)
plt.bar(models, recall_scores,width=0.5)
plt.title('Recall trung bình của các mô hình')
plt.ylabel('Recall (%)')
plt.ylim([0, 100])

# Ghi chú các giá trị phần trăm lên biểu đồ
for i, v in enumerate(recall_scores):
    plt.annotate(f'{v:.2f}%', xy=(i, v), ha='center', va='top')

# Biểu đồ độ chính xác (dưới)
plt.subplot(2, 2, 3)
plt.bar(models, precision_scores,width=0.5)
plt.title('Precision trung bình của các mô hình')
plt.ylabel('Precision (%)')
plt.ylim([0, 100])

# Ghi chú các giá trị phần trăm lên biểu đồ
for i, v in enumerate(precision_scores):
    plt.annotate(f'{v:.2f}%', xy=(i, v), ha='center', va='top')

# Biểu đồ F1-score (dưới)
plt.subplot(2, 2, 4)
plt.bar(models, f1_scores,width=0.5)
plt.title('F1-score trung bình của các mô hình')
plt.ylabel('F1-score (%)')
plt.ylim([0, 100])

# Ghi chú các giá trị phần trăm lên biểu đồ
for i, v in enumerate(f1_scores):
    plt.annotate(f'{v:.2f}%', xy=(i, v), ha='center', va='top')

plt.tight_layout()
plt.show()
```
